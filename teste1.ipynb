```python
import plotly.express as px

# Suponha que você tenha duas listas numéricas, por exemplo:
x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 18, 20]

# Crie um DataFrame com os dados
import pandas as pd
data = pd.DataFrame({'x': x, 'y': y})

# Use o Plotly Express para criar um gráfico de dispersão
fig = px.scatter(data, x='x', y='y', title='Gráfico de Dispersão')

# Exiba o gráfico
fig.show("png")
```
